```python
from PIL import Image
import os
import pandas as pd
from ydata_profiling import ProfileReport
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
minitest_path = '/content/drive/MyDrive/minitest'

```

```python
def analyze_images(root_path, file_extension='.jpeg'):
    data = []
    for filename in os.listdir(root_path):
        if filename.lower().endswith(file_extension):
            try:
                image_path = os.path.join(root_path, filename)
                file_size = os.path.getsize(image_path)  # Get the file size in bytes
                with Image.open(image_path) as img:
                    width, height = img.size
                    dimensions = f"{width} x {height}"
                    data.append({'Filename': filename,
                                 'Dimensions': dimensions,
                                 'File Size (bytes)': file_size})
            except Exception as e:
                print(f"Error processing file {filename}: {e}")
    return pd.DataFrame(data)



```

```python
minitest_data = analyze_images(minitest_path)

```
