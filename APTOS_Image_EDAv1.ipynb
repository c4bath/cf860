```python
from PIL import Image
import os
import pandas as pd
from ydata_profiling import ProfileReport
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_path = '/content/drive/MyDrive/aptos'
train_path = os.path.join(base_path, 'train')
test_path = os.path.join(base_path, 'test')
valid_path = os.path.join(base_path, 'valid')
```

```python
def analyze_images(root_path, file_extension='.png'):
    data = []
    for subdir, dirs, files in os.walk(root_path):
        for filename in files:
            if filename.lower().endswith(file_extension):
                try:
                    image_path = os.path.join(subdir, filename)
                    file_size = os.path.getsize(image_path)  # Get the file size in bytes
                    with Image.open(image_path) as img:
                        width, height = img.size
                        dimensions = f"{width} x {height}"
                        data.append({'Filename': filename,
                                     'Dimensions': dimensions,
                                     'File Size (bytes)': file_size,
                                     'Subdirectory': subdir})
                except Exception as e:
                    print(f"Error processing file {filename} in {subdir}: {e}")
    return pd.DataFrame(data)



```

```python
train_data = analyze_images(train_path)
test_data = analyze_images(test_path)
valid_data = analyze_images(valid_path)
```
